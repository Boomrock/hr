```python
# –°–æ–∑–¥–∞–Ω–∏–µ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('–ê–Ω–∞–ª–∏–∑ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π HR Chat Companion', fontsize=16, fontweight='bold')

# –ì—Ä–∞—Ñ–∏–∫ 1: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
axes[0, 0].pie(fr_categories.values, labels=fr_categories.index, autopct='%1.1f%%', startangle=90)
axes[0, 0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')

# –ì—Ä–∞—Ñ–∏–∫ 2: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º
colors = {'–í—ã—Å–æ–∫–∏–π': '#ff6b6b', '–°—Ä–µ–¥–Ω–∏–π': '#feca57', '–ù–∏–∑–∫–∏–π': '#48cae4'}
priority_colors = [colors.get(p, '#95a5a6') for p in fr_priorities.index]
axes[0, 1].bar(fr_priorities.index, fr_priorities.values, color=priority_colors)
axes[0, 1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º')
axes[0, 1].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π')

# –ì—Ä–∞—Ñ–∏–∫ 3: –°—Ç–∞—Ç—É—Å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏
status_colors = {'–†–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ': '#2ecc71', '–ß–∞—Å—Ç–∏—á–Ω–æ': '#f39c12', '–ù–µ —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ': '#e74c3c'}
status_plot_colors = [status_colors.get(s, '#95a5a6') for s in fr_status.index]
axes[1, 0].bar(fr_status.index, fr_status.values, color=status_plot_colors)
axes[1, 0].set_title('–°—Ç–∞—Ç—É—Å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏')
axes[1, 0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π')

# –ì—Ä–∞—Ñ–∏–∫ 4: Heatmap –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ vs –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç—ã
priority_category = pd.crosstab(fr_df['category'], fr_df['priority'])
sns.heatmap(priority_category, annot=True, fmt='d', cmap='YlOrRd', ax=axes[1, 1])
axes[1, 1].set_title('–ö–∞—Ç–µ–≥–æ—Ä–∏–∏ vs –ü—Ä–∏–æ—Ä–∏—Ç–µ—Ç—ã')

plt.tight_layout()
plt.show()

print("üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π —Å–æ–∑–¥–∞–Ω–∞")
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω–µ—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('–ê–Ω–∞–ª–∏–∑ –Ω–µ—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π HR Chat Companion', fontsize=16, fontweight='bold')

# –ì—Ä–∞—Ñ–∏–∫ 1: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
axes[0].pie(nfr_categories.values, labels=nfr_categories.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')

# –ì—Ä–∞—Ñ–∏–∫ 2: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º
nfr_priority_colors = [colors.get(p, '#95a5a6') for p in nfr_priorities.index]
axes[1].bar(nfr_priorities.index, nfr_priorities.values, color=nfr_priority_colors)
axes[1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º')
axes[1].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π')

# –ì—Ä–∞—Ñ–∏–∫ 3: –°—Ç–∞—Ç—É—Å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏
nfr_status_colors = [status_colors.get(s, '#95a5a6') for s in nfr_status.index]
axes[2].bar(nfr_status.index, nfr_status.values, color=nfr_status_colors)
axes[2].set_title('–°—Ç–∞—Ç—É—Å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏')
axes[2].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π')

plt.tight_layout()
plt.show()

print("üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω–µ—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π —Å–æ–∑–¥–∞–Ω–∞")
```

```python
# üéØ –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–≤

## –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –≤—Å–µ—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π
all_requirements = pd.concat([
    fr_df.assign(type='–§—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ'),
    nfr_df.assign(type='–ù–µ—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ')
])

# –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–æ–Ω–∫—É —Å ID —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è
all_requirements['requirement_id'] = all_requirements.index

# –ü–µ—Ä–µ—É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–µ–º –∫–æ–ª–æ–Ω–∫–∏ –¥–ª—è –ª—É—á—à–µ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
columns_order = ['requirement_id', 'type', 'name', 'category', 'priority', 'status', 'description']
all_requirements = all_requirements[columns_order]

print("üìã –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –≤—Å–µ—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π:")
print(f"–í—Å–µ–≥–æ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π: {len(all_requirements)}")
print(f"–§—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö: {len(all_requirements[all_requirements['type'] == '–§—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ'])}")
print(f"–ù–µ—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã—Ö: {len(all_requirements[all_requirements['type'] == '–ù–µ—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ'])}")

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –ø–µ—Ä–≤—ã—Ö 10 —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π
all_requirements.head(10)
```

```python
# –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç—É
print("üîç –§–∏–ª—å—Ç—Ä –ø–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º:")
print("\n" + "="*50)

for priority in ['–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π', '–í—ã—Å–æ–∫–∏–π', '–°—Ä–µ–¥–Ω–∏–π', '–ù–∏–∑–∫–∏–π']:
    filtered = all_requirements[all_requirements['priority'] == priority]
    if not filtered.empty:
        print(f"\nüìå {priority} –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç ({len(filtered)} —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π):")
        for idx, row in filtered.iterrows():
            print(f"  ‚Ä¢ {idx}: {row['name']}")

# –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Å—Ç–∞—Ç—É—Å—É —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏
print("\n" + "="*50)
print("‚úÖ –°—Ç–∞—Ç—É—Å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –ø–æ —Ç–∏–ø–∞–º —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π:")

status_by_type = pd.crosstab(all_requirements['type'], all_requirements['status'], margins=True)
print(status_by_type)
```

```python
# üìù –í—ã–≤–æ–¥—ã –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏

## –ê–Ω–∞–ª–∏–∑ —Ç–µ–∫—É—â–µ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è –ø—Ä–æ–µ–∫—Ç–∞
```

```python
# –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫ –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏ –ø—Ä–æ–µ–∫—Ç–∞
total_requirements = len(all_requirements)
implemented = len(all_requirements[all_requirements['status'] == '–†–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ'])
partially_implemented = len(all_requirements[all_requirements['status'] == '–ß–∞—Å—Ç–∏—á–Ω–æ'])

completion_rate = (implemented / total_requirements) * 100
partial_completion_rate = ((implemented + partially_implemented) / total_requirements) * 100

print("üìä –ú–ï–¢–†–ò–ö–ò –ì–û–¢–û–í–ù–û–°–¢–ò –ü–†–û–ï–ö–¢–ê")
print("="*50)
print(f"üìà –û–±—â–∞—è –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—å: {completion_rate:.1f}%")
print(f"üìà –ì–æ—Ç–æ–≤–Ω–æ—Å—Ç—å —Å —É—á–µ—Ç–æ–º —á–∞—Å—Ç–∏—á–Ω–æ–π —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏: {partial_completion_rate:.1f}%")
print(f"‚úÖ –ü–æ–ª–Ω–æ—Å—Ç—å—é —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ: {implemented} –∏–∑ {total_requirements}")
print(f"üîÑ –ß–∞—Å—Ç–∏—á–Ω–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ: {partially_implemented} –∏–∑ {total_requirements}")

# –ê–Ω–∞–ª–∏–∑ –ø–æ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–º —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º
critical_requirements = all_requirements[all_requirements['priority'] == '–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π']
critical_implemented = len(critical_requirements[critical_requirements['status'] == '–†–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ'])
critical_total = len(critical_requirements)

if critical_total > 0:
    critical_completion = (critical_implemented / critical_total) * 100
    print(f"\nüö® –ö–†–ò–¢–ò–ß–ï–°–ö–ò–ï –¢–†–ï–ë–û–í–ê–ù–ò–Ø:")
    print(f"   –ì–æ—Ç–æ–≤–Ω–æ—Å—Ç—å: {critical_completion:.1f}% ({critical_implemented}/{critical_total})")

# –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏
print(f"\nüí° –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò:")
print("="*50)

if completion_rate >= 80:
    print("üéâ –û—Ç–ª–∏—á–Ω–∞—è —Ä–∞–±–æ—Ç–∞! –ü—Ä–æ–µ–∫—Ç –±–ª–∏–∑–æ–∫ –∫ –∑–∞–≤–µ—Ä—à–µ–Ω–∏—é.")
elif completion_rate >= 60:
    print("üëç –•–æ—Ä–æ—à–∏–π –ø—Ä–æ–≥—Ä–µ—Å—Å! –ü—Ä–æ–¥–æ–ª–∂–∞–π—Ç–µ —Ä–∞–±–æ—Ç—É –Ω–∞–¥ –æ—Å—Ç–∞–≤—à–∏–º–∏—Å—è —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º–∏.")
elif completion_rate >= 40:
    print("‚ö†Ô∏è  –°—Ä–µ–¥–Ω–∏–π –ø—Ä–æ–≥—Ä–µ—Å—Å. –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è —Å–æ—Å—Ä–µ–¥–æ—Ç–æ—á–∏—Ç—å—Å—è –Ω–∞ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã—Ö —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è—Ö.")
else:
    print("üî¥ –ù–∏–∑–∫–∏–π —É—Ä–æ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏. –¢—Ä–µ–±—É–µ—Ç—Å—è –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–∞—è —Ä–∞–±–æ—Ç–∞ –Ω–∞–¥ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–µ–π.")

# –°–ª–µ–¥—É—é—â–∏–µ —à–∞–≥–∏
high_priority_pending = all_requirements[
    (all_requirements['priority'] == '–í—ã—Å–æ–∫–∏–π') & 
    (all_requirements['status'] != '–†–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ')
]

if not high_priority_pending.empty:
    print(f"\nüéØ –ü–†–ò–û–†–ò–¢–ï–¢–ù–´–ï –ó–ê–î–ê–ß–ò ({len(high_priority_pending)}):")
    for idx, row in high_priority_pending.iterrows():
        print(f"   ‚Ä¢ {idx}: {row['name']}")
```
